```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_table("D:\Professional\Projects\Data Structures in C\Amortized Analysis for Dynamic Table\Data.txt")
```

```python
df
```

```python
#Set the figure size:
plt.figure(figsize = (14,10))

#Plot the line graphs for varying y and constant x:
plt.plot(df['i'], df['Size_i'], linestyle = "dashed", label = "Size_i")
plt.plot(df['i'], df['Num_i'], label = "Num_i", linewidth = 0.75)
plt.plot(df['i'], df['C_i'], label = "Cost_i")
plt.plot(df['i'], df['Phi_i'], label = "Phi_i", linewidth = 3)
plt.plot(df['i'], df["C_i'"], label = "Cost_i'", marker = 'D', markerfacecolor = "black", markersize = 2.5)

#Set the aesthetics:
plt.title("Change of different parameters w.r.t. i and the ith operation")
plt.xlabel("i")
plt.ylabel("Unit value")
plt.legend()

#Show the plot:
plt.show()
```
